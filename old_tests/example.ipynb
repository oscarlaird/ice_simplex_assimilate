```python
mat = loadmat('./example_data/icepack2.mat')
prior_a = mat['prior_aicen'].T
prior_v = mat['prior_vicen'].T
true_a = mat['true_aicen'].T
h_bnd = ice.deltize.HeightBounds([0, 0.64450721681942580, 1.3914334975763036, 2.4701793819598885, 4.5672879188504911, 9.3338418158681744])
```

```python
i = 0
a0, v0 = prior_a[i], prior_v[i]
raw_ensemble = ice.deltize.build_raw_ensemble(a0, v0)
samples = ice.deltize.process_ensemble(raw_ensemble, h_bnd)
x_0 = np.ones(len(samples)) * (1 - true_a[i].sum())
out = sa.transport.transport_pipeline(samples, x_0)
out
```
